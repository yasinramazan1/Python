```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
liste1 = [10,15,20,25,30,35,40,45,50]
liste2 = [123,346,235,456,254,567,857,576,234]
```

```python
array1 = np.array(liste1)
array2 = np.array(liste2)
```

```python
plt.plot(array1, array2, "green")
```

```python
plt.plot(array1, array2, "blue")
```

```python
plt.title("Grafik Başlığı")
plt.plot(array1, array2, "blue")
```

```python
plt.plot(array1, array2, "blue")
plt.title("Python Matplotlib Kütüphanesi Çalışması")
plt.xlabel("İndeksler")
plt.ylabel("Değerler")
```

```python
plt.show() # bu komut ile bu grafik ile ilgili çalışmayı bitiririz ve aşağıda yazacağımız kodlar bu grafiği etkilemeyecektir.
```

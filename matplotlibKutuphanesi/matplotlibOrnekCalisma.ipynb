```python
x = range(1,44)
y = data["Unit Price"]
plt.figure(figsize=(20,10))
plt.xlabel("Günler")
plt.ylabel("Fiyatlar")
plt.title("Satış Tablosu")
plt.xticks(np.arange(1,44,1))
plt.bar(x,y, color="green", width=0.8)
plt.plot(x,y,color="red")
plt.show()
```

```python
x = range(1,44)
y = data["Unit Price"]
plt.figure(figsize=(20,10))
plt.xlabel("Günler")
plt.ylabel("Fiyatlar")
plt.title("Satış Tablosu")
plt.xticks(np.arange(1,44,1))
plt.bar(x,y, color="green", width=0.8)
plt.plot(x,y,color="red")
plt.legend(labels=["Satışlar"])
plt.show()
```

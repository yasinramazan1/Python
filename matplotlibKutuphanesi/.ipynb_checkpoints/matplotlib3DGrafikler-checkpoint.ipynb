```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

z=np.random.rand(10,3)
x = z*np.random.rand(10,3)
y = x*np.random.rand(10,3)

ax.plot_surface(x,y,z)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

z=np.random.rand(10,3)
x = z*np.random.rand(10,3)
y = x*np.random.rand(10,3)

ax.plot_surface(x,y,z);
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

z=np.random.rand(10,3)
x = z*np.random.rand(10,3)
y = x*np.random.rand(10,3)

ax.plot_surface(x,y,z);
plt.title("3D Grafik")
plt.show()
```

```python
fig = plt.figure(figsize=(10,10)) # Grafik boyutunu belirler.
ax = plt.axes(projection="3d")

z=np.random.rand(10,3)
x = z*np.random.rand(10,3)
y = x*np.random.rand(10,3)

ax.plot_surface(x,y,z);
plt.title("3D Grafik")
plt.show()
```

```python
fig = plt.figure(figsize=(8,8)) # Grafik boyutunu belirler.
ax = plt.axes(projection="3d")

z=np.random.rand(10,3)
x = z*np.random.rand(10,3)
y = x*np.random.rand(10,3)

ax.plot_surface(x,y,z, color="green");
plt.title("3D Grafik")
plt.show()
```

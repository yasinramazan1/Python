```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
x=[14,13,34,54]
y=["a","b","c","d"]
plt.pie(x, labels= y)
```

```python
plt.figure(figsize=(8,8))
x=[14,13,34,54]
y=["a","b","c","d"]
plt.pie(x, labels= y);
```

```python
plt.figure(figsize=(8,8))
x=[14,13,34,54]
y=["a","b","c","d"]
plt.pie(x, labels= y);
plt.show()
```

```python
plt.figure(figsize=(8,8))
x=[14,13,34,54]
y=["a","b","c","d"]
e=[0.3,0,0,0]
plt.pie(x, explode=e,labels= y);
plt.show()
```

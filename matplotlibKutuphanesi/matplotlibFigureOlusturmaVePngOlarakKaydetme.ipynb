```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dizi1 = np.linspace(1,10,10)
dizi1
```

```python
dizi2=dizi1**2
dizi2
```

```python
fig1=plt.figure() # 0 axes demek, eksenleri yok demektir.
```

```python
fig1=plt.figure()
fig2=fig1.add_axes([0.2,0.2,1,1]) # başka grafiğe göre yatay konum, düşey konum, genişlik, yükseklik
```

```python
fig1=plt.figure()
fig2=fig1.add_axes([0.2,0.2,1,1])
fig2.plot(dizi1, dizi2, "r")
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
ax1=fig.add_axes([0.1,0.1,0.7,0.7])
ax2=fig.add_axes([0.1,0.1,0.3,0.3])

ax1.plot(dizi1, dizi2, "red")
ax2.plot(dizi1,dizi2, "blue") 
ax1.set_title("Kırmızı Başlık")
ax2.set_title("Mavi Başlık")
ax1.set_xlabel("Kırmızı x Ekseni")
ax1.set_ylabel("Kırmızı y ")
ax2.set_xlabel("Mavi x Ekseni")
ax2.set_ylabel("Mavi y ")
```

```python
fig = plt.figure()
ax1=fig.add_axes([0.1,0.1,0.7,0.7])
ax2=fig.add_axes([0.1,0.4,0.3,0.3]) # Konum değiştirme

ax1.plot(dizi1, dizi2, "red")
ax2.plot(dizi1,dizi2, "blue") 
ax1.set_title("Kırmızı Başlık")
ax2.set_title("Mavi Başlık")
ax1.set_xlabel("Kırmızı x Ekseni")
ax1.set_ylabel("Kırmızı y ")
ax2.set_xlabel("Mavi x Ekseni")
ax2.set_ylabel("Mavi y ")
```

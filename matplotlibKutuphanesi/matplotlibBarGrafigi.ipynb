```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
liste = ["X", "Y", "Z", "T"]
liste2 = [10,20,15,13]
plt.bar(liste, liste2)
plt.show()
```

```python
liste = ["X", "Y", "Z", "T"]
liste2 = [10,20,15,13]
plt.bar(liste, liste2,color="yellow")
plt.show()
```

```python
liste = ["X", "Y", "Z", "T"]
liste2 = [10,20,15,13]
plt.bar(liste, liste2,color="yellow", width=0.5)
plt.show()
```

```python
liste = ["X", "Y", "Z", "T"]
liste2 = [10,20,15,13]
plt.barh(liste, liste2,color="yellow", height=0.5)
plt.show()
```

```python
liste3= [18,15,7,5]
liste4= [6,9,5,2]
x=np.arange(4)

plt.bar(x, liste3, color="purple", width=0.3)
plt.bar(x, liste4, color="green", width=0.6)
plt.xlabel("Yatay")
plt.ylabel("Dikey")
plt.show()
```

```python
liste3= [18,15,7,5]
liste4= [6,9,5,2]
x=np.arange(4)

plt.bar(x, liste3, color="purple", width=0.3)
plt.bar(x, liste4, color="green", width=0.6)
plt.xlabel("Yatay")
plt.ylabel("Dikey")
plt.xticks(x,("X", "Y", "Z", "T"))
plt.show()
```

```python
liste3= [18,15,7,5]
liste4= [6,9,5,2]
x=np.arange(4)

plt.bar(x, liste3, color="purple", width=0.3)
plt.bar(x, liste4, color="green", width=0.6)
plt.xlabel("Yatay")
plt.ylabel("Dikey")
plt.xticks(x,("X", "Y", "Z", "T"))
plt.legend(labels=["Mor","Yeşil"])
plt.show()
```

```python
liste3= [18,15,7,5]
liste4= [6,9,5,2]
x=np.arange(4)

plt.bar(x, liste3, color="purple", width=0.3)
plt.bar(x, liste4, color="green", width=0.6)
plt.xlabel("Yatay")
plt.ylabel("Dikey")
plt.yticks(np.arange(0,25,1))
plt.xticks(x,("X", "Y", "Z", "T"))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.random.rand(5,10)*100
y = np.random.randn(5,10)*100
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.rand(15,10)*100
y = np.random.randn(15,10)*100
plt.scatter(x,y)

z = np.random.rand(20,10)*50
t = np.random.randn(20,10)*50
plt.scatter(z,t)
plt.show()
```

```python
x = np.random.rand(15,10)*100
y = np.random.randn(15,10)*100
plt.scatter(x,y, color="yellow")

z = np.random.rand(20,10)*50
t = np.random.randn(20,10)*50
plt.scatter(z,t,color="purple")
plt.show()
```

```python
x = np.random.rand(15,10)*100
y = np.random.randn(15,10)*100
plt.scatter(x,y, color="yellow")

z = np.random.rand(20,10)*50
t = np.random.randn(20,10)*50
plt.scatter(z,t,color="black",alpha=0.5) # alpha saydaml ayarlar.
plt.show() 
```

```python
array1=np.array([1,3,5,7,9])
array2=np.array([10,30,50,70,90])

renk = ["yellow", "purple", "blue", "red", "green"]
plt.scatter(array1, array2, color=renk)
plt.show()
```

```python
array1=np.array([1,3,5,7,9])
array2=np.array([10,30,50,70,90])

renk = ["yellow", "purple", "blue", "red", "green"]
plt.scatter(array1, array2, color=renk)
plt.colorbar()
plt.show()
```

```python
array1=np.array([1,3,5,7,9])
array2=np.array([10,30,50,70,90])

renk = ["yellow", "purple", "blue", "red", "green"]
dimens = [250,750,1250,2500,4000]
plt.scatter(array1, array2, color=renk, s=dimens) # s=size'dan gelir.
plt.colorbar()
plt.show()
```
